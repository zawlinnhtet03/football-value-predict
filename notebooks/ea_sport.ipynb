```python
# Impute missing values for goalkeepers with the mean of their stats
goalkeeper_stats_mean = df[['Pace', 'Shooting', 'Passing', 'Dribbling', 'Defending', 'Physic']].mean()
df.loc[df['Positions'] == 'GK', ['Pace', 'Shooting', 'Passing', 'Dribbling', 'Defending', 'Physic']] = df.loc[df['Positions'] == 'GK', ['Pace', 'Shooting', 'Passing', 'Dribbling', 'Defending', 'Physic']].fillna(goalkeeper_stats_mean)

df.isnull().sum()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map

```

```python
nationality_map = shorten_categories(df['Nationality'].value_counts(), 1000)
df['Nationality'] = df['Nationality'].map(nationality_map)
df['Nationality'].unique()
```

```python
le_foot = LabelEncoder()
df['Preferred_foot'] = le_foot.fit_transform(df['Preferred_foot'])
```

```python
le_experience = LabelEncoder()
df['experience'] = le_experience.fit_transform(df['experience'])
```

```python
X = df[['Nationality', 'Primary_Position', 'Age', 'Pace', 'Shooting', 'Passing', 'Dribbling', 'Defending', 'Physic', 'Goalkeeping', 'Overall']]
y = df['Value_euro']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Train Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(X, y)
y_pred_lr = linear_reg.predict(X)
error_lr = np.sqrt(mean_squared_error(y, y_pred_lr))
print(f"Linear Regression RMSE: {error_lr:.2f}")

# Train Decision Tree Regressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X, y)
y_pred_dt = dec_tree_reg.predict(X)
error_dt = np.sqrt(mean_squared_error(y, y_pred_dt))
print(f"Decision Tree RMSE: {error_dt:.2f}")

# Train Random Forest Regressor
random_forest_reg = RandomForestRegressor(n_estimators=50, n_jobs=-1, random_state=0)
random_forest_reg.fit(X, y)
y_pred_rf = random_forest_reg.predict(X)
error_rf = np.sqrt(mean_squared_error(y, y_pred_rf))
print(f"Random Forest RMSE: {error_rf:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Example with Decision Tree
param_grid = {"max_depth": [None, 5, 10, 15, 20]}
grid_search = GridSearchCV(DecisionTreeRegressor(random_state=0), param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X, y)
best_tree = grid_search.best_estimator_
print("Best Decision Tree RMSE: ", np.sqrt(mean_squared_error(y, best_tree.predict(X))))
```
